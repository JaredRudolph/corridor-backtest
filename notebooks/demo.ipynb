# Portallocator Backtest Demo

End-to-end walkthrough: fetch prices, run backtest, inspect results.

```python
import sys
sys.path.insert(0, '..')

import matplotlib.pyplot as plt

from config import config
from allocator.pipeline import run_pipeline
```

```python
output = run_pipeline(config)
results = output['results']
metrics = output['metrics']
events = output['events']
```

```python
p = metrics['portfolio']
b = metrics['benchmark']

print(f"{'Metric':<20} {'Portfolio':>12} {'Benchmark':>12}")
print('-' * 46)
print(f"{'CAGR':<20} {p['cagr']:>12.2%} {b['cagr']:>12.2%}")
print(f"{'Sharpe':<20} {p['sharpe']:>12.2f} {b['sharpe']:>12.2f}")
print(f"{'Max Drawdown':<20} {p['max_drawdown']:>12.2%} {b['max_drawdown']:>12.2%}")
print(f"{'Final Value':<20} ${p['final_value']:>11,.0f}")
print(f"\nRebalance events:    {metrics['rebalance_events']}")
print(f"Total contributions: ${metrics['total_contributions']:,.0f}")
print(f"Growth:              ${metrics['growth']:,.0f}")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

results['portfolio_value'].plot(ax=axes[0], label='Portfolio', color='steelblue', linewidth=1.5)
axes[0].set_title('Portfolio Value')
axes[0].set_ylabel('Value ($)')
axes[0].legend()

weight_cols = [c for c in results.columns if c.startswith('weight_')]
results[weight_cols].rename(columns=lambda c: c.replace('weight_', '')).plot(ax=axes[1])
axes[1].set_title('Asset Weights Over Time')
axes[1].set_ylabel('Weight')
axes[1].set_ylim(0, 1)

rebal_dates = results.index[results['rebalanced']]
for ax in axes:
    for d in rebal_dates:
        ax.axvline(d, color='red', alpha=0.2, linewidth=0.5)

plt.tight_layout()
plt.show()
```
